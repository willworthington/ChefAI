# Evaluating GPT Based on Human Scores

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
scores = pd.read_csv('/content/drive/MyDrive/Data_Lab_Final_Project/Evaluation/GPT_All_Evaluations.csv')
scores = scores.dropna()
scores
```

```python
all_scores = []
for col in scores.columns:
  all_scores.extend(scores[col].tolist())

print("Total scores recorded:", len(all_scores))
```

```python
scores_np = scores.to_numpy()
mean_scores = np.mean(scores_np, axis=1)
best_r = np.argmax(mean_scores)
worst_r = np.argmin(mean_scores)

print('Best recipe: '+str(best_r))
print('Worst recipe: '+str(worst_r))
mean_scores[worst_r]
```

```python
plt.hist(x=all_scores, bins=10, density=True)
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('Rate')
plt.show()
```

```python
print("Mean Score:", np.mean(all_scores))
```

```python
print("Score Variance:", np.std(all_scores))
```
