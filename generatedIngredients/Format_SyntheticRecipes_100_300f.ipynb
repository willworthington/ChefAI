```python
# Create a dict with unique key-value pairs of ingredient id to ingredient
ingredientDict = {}

for row in range(len(ingKey)):
    ingredient = ingKey.at[row, 'replaced']
    id_code = ingKey.at[row, 'id']
    
    if id_code not in ingredientDict:
        ingredientDict[id_code] = ingredient
```

```python
decoded_recipes = []
count = 0

cols = recipes.columns

for i in range(len(recipes.index)):
    currentRecipe = []
    for j in range(len(recipes.columns)):
        if recipes.at[i, cols[j]] == True:
            identifier = int(cols[j].split("_")[1])
            currentRecipe.append(ingredientDict[identifier])
    decoded_recipes.append(currentRecipe)
```

```python
# Decoded_recipes holds the ingredients list for each of the 100 recipes
for recipe in decoded_recipes:
    print(recipe)
```
